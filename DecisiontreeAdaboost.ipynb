```
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd 
```

```
data = pd.read_csv("/content/drive/My Drive/Minor2/shuffled_dataset.csv") 
data.head()
```

```
import numpy as np
X=np.array(data.iloc[:,3:32403].values)
Y=data.iloc[:,2].values
print(Y)
print(X)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```
from sklearn.tree import DecisionTreeClassifier
classifier = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=3),
    n_estimators=100
)
decisada = classifier.fit(X_train, Y_train)
```

```
predictions = classifier.predict(X_test)
```

```
print("Accuracy:",metrics.accuracy_score(Y_test, predictions))
```

```
import pickle

with open('/content/drive/My Drive/Minor2/decisionada.pck', 'wb') as f:
    pickle.dump(decisada,f)
```
